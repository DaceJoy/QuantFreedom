```python
import quantfreedom as qf
from quantfreedom.data.data_dl import data_download_from_ccxt
import pandas as pd
import numpy as np
import ccxt

%load_ext autoreload
%autoreload 2
```

```python
exchange = ccxt.binance()
exchange.load_markets()
exchange.symbols
```

```python
exchange.parse8601('2023-09-09T00:00:00Z')
```

```python
all_ohlcvs = exchange.fetch_ohlcv(
    symbol="BTC/USDT",
    timeframe="4h",
    since=exchange.parse8601("2023-09-11T00:00:00Z"),
)
```

```python
all_ohlcvs = np.array(all_ohlcvs)
data_columns = pd.MultiIndex.from_tuples(
    [
        ('BTCUSDT', "open"),
        ('BTCUSDT', "high"),
        ('BTCUSDT', "low"),
        ('BTCUSDT', "close"),
        ('BTCUSDT', "volume"),
    ],
    names=["symbol", "candle_info"],
)
data_index = pd.Index(
    data=pd.to_datetime(all_ohlcvs[:, 0].flatten(), unit="ms"),
    name="open_time",
)
data_btc = pd.DataFrame(
    all_ohlcvs[:, 1:],
    columns=data_columns,
    index=data_index,
)
data_btc.drop(columns=('BTCUSDT', "volume"), inplace=True, axis=1)
data_btc
```

```python
all_ohlcvs = np.array(exchange.fetch_ohlcv(
    symbol="ETH/USDT",
    timeframe="4h",
    since=exchange.parse8601("2023-09-11T00:00:00Z"),
))
data_columns = pd.MultiIndex.from_tuples(
    [
        ('ETHUSDT', "open"),
        ('ETHUSDT', "high"),
        ('ETHUSDT', "low"),
        ('ETHUSDT', "close"),
        ('ETHUSDT', "volume"),
    ],
    names=["symbol", "candle_info"],
)
data_index = pd.Index(
    data=pd.to_datetime(all_ohlcvs[:, 0].flatten(), unit="ms"),
    name="open_time",
)
data_eth = pd.DataFrame(
    all_ohlcvs[:, 1:],
    columns=data_columns,
    index=data_index,
)
data_eth.drop(columns=('ETHUSDT', "volume"), inplace=True, axis=1)
data_eth
```

```python
data = data_btc.join(data_eth)
```
