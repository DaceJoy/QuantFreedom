```python
import numpy as np
import pandas as pd
from quantfreedom.testing import testing_combine_evals
from quantfreedom.evaluators import eval_is_below
from quantfreedom.indicators.talib_ind import from_talib
from itertools import product
prices = pd.read_hdf('../tests/data/prices.hd5')
```

```python
rsi_ind = from_talib(
    func_name='rsi',
    prices=prices,
    cart_product=False,
    combos=False,
    timeperiod=[15, 30],
)
ema_ind = from_talib(
    func_name='ema',
    prices=prices,
    cart_product=False,
    combos=False,
    timeperiod=[100, 200, 300],
)
rsi_ema_ind = from_talib(
    func_name='ema',
    indicator_data=rsi_ind,
    cart_product=False,
    combos=False,
    timeperiod=[100, 200, 300],
)
```

```python
ema_is_below = eval_is_below(
    want_to_evaluate=ema_ind,
    prices=prices,
    cand_ohlc='close'
)
rsi_is_below = eval_is_below(
    want_to_evaluate=rsi_ind,
    user_args=[50,60],
)
rsi_ema_is_below = eval_is_below(
    want_to_evaluate=rsi_ema_ind,
    indicator_data=rsi_ind,
)
```

```python
testing_combine_evals(
    bigger_df=rsi_is_below,
    smaller_df=rsi_ema_is_below,
)['BTCUSDT']
```

```python
testing_combine_evals(
    bigger_df=ema_is_below,
    smaller_df=rsi_is_below,
)['BTCUSDT']
```
