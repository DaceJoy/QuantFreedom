```python
raw_tick_data = pd.read_csv("C:/Users/User/Desktop/BTCUSDT-aggTrades-2023-10-14.csv").drop(
    ["agg_trade_id", "first_trade_id", "last_trade_id"], axis=1
)
raw_tick_data["datetime"] = pd.to_datetime(raw_tick_data.transact_time, unit="ms")
raw_tick_data.set_index("datetime", inplace=True)
raw_tick_data["is_buyer_maker"].replace([False, True], ["buy", "sell"], inplace=True)
raw_tick_data["usdt"] = raw_tick_data["price"] * raw_tick_data["quantity"]
raw_tick_data.rename(columns={"is_buyer_maker": "side", "transact_time": "timestamp", "quantity": "qty"}, inplace=True)
raw_tick_data["price"] = raw_tick_data['price']
raw_tick_data["usdt"] = raw_tick_data['usdt'].round(4)
tick_data = raw_tick_data[["timestamp", "price", "usdt", "side"]]
tick_list = tick_data.values[:300].tolist()
tick_list
```
